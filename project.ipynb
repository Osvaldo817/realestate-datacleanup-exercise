#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# this CSV file contains semicolons instead of comas as separator
ds = pd.read_csv('assets/real_estate.csv', sep=';')

selected_populations = ['Fuenlabrada', 'Leganes', 'Getafe', 'Alcorcon']

subset_df = ds[ds['level5'].isin(selected_populations)]

normalized_prices = subset_df.groupby('level5')['price'].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

plt.figure(figsize=(10, 6))

for population, prices in normalized_prices.groupby('level5'):
    plt.hist(prices, bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices for Selected Populations')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file and create a subset DataFrame for the selected towns
ds = pd.read_csv('assets/real_estate.csv', sep=';')
selected_populations = ['Fuenlabrada', 'Leganes', 'Getafe', 'Alcorcon']
subset_df = ds[ds['level5'].isin(selected_populations)]

# Calculate price per square meter (price/m2)
subset_df['pps'] = subset_df['price'] / subset_df['surface']

# Filter data for Getafe and Alcorcon
getafe_data = subset_df[subset_df['level5'] == 'Getafe']
alcorcon_data = subset_df[subset_df['level5'] == 'Alcorcon']

# Plotting
plt.scatter(getafe_data['pps'], getafe_data['price'], label='Getafe', color='Blue')
plt.scatter(alcorcon_data['pps'], alcorcon_data['price'], label='Alcorcon', color='Red')
plt.xlabel('Price per Square Meter (Price/m2)')
plt.ylabel('Total Price')
plt.legend()
plt.title('Price per Square Meter Comparison between Getafe and Alcorcon')
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# this CSV file contains semicolons instead of comas as separator
ds = pd.read_csv('assets/real_estate.csv', sep=';')

selected_populations = ['Fuenlabrada', 'Leganes', 'Getafe', 'Alcorcon']

subset_df = ds[ds['level5'].isin(selected_populations)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Scatter Plots for Selected Populations')

for i, (population, data) in enumerate(subset_df.groupby('level5')):
    row, col = divmod(i, 2)
    axes[row, col].scatter(data['surface'], data['price'], label=population, alpha=0.7)
    axes[row, col].set_title(population)
    axes[row, col].set_xlabel('surface')
    axes[row, col].set_ylabel('price')
    axes[row, col].legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
